Task 2: Change color spaces (RGB -> Grayscale, HSV, LAB)

```python
img_2 = cv2.imread('input_openCV.jpg', cv2.IMREAD_COLOR)

img_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img_2, cv2.COLOR_BGR2HSV)
img_lab = cv2.cvtColor(img_2, cv2.COLOR_BGR2LAB)
```

```python
for img in [img_gray, img_hsv, img_lab]:
    cv2.imshow('Task2: Change color spaces', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Task 3: Crop and resize images

```python
img_3a = cv2.imread('input_openCV.jpg', cv2.IMREAD_COLOR)

print("Type:", type(img_3a))
print("Image shape:", img_3a.shape)

crop = img_3a[100:300, 100:300]
cv2.imshow("Task 3: Crop", crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

img_3b = cv2.imread('input_openCV.jpg', cv2.IMREAD_COLOR)

basic = cv2.resize(img_3b, (1200, 675))
small = cv2.resize(img_3b, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_AREA)
large = cv2.resize(img_3b, (1050, 1610), interpolation=cv2.INTER_CUBIC)
medium = cv2.resize(img_3b, (780, 540), interpolation=cv2.INTER_LINEAR)

titles = ["Original", "10% (INTER_AREA)", "1050x1610 (INTER_CUBIC)", "780x540 (INTER_LINEAR)"]
images = [basic, small, large, medium]

plt.figure(figsize=(10, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)  
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

Task 4: Draw shape and text on images

```python
img_4 = cv2.imread('input_openCV.jpg', cv2.IMREAD_COLOR)

cv2.line(img_4, (50, 50), (200, 200), (0, 255, 0), thickness=5)
cv2.rectangle(img_4, (300, 50), (450, 200), (255, 0, 0), thickness=5)
cv2.circle(img_4, (600, 125), 75, (0, 0, 255), thickness=5)

txt = "Computer Vision Lab 1 With Ms Tien"
cv2.putText(img_4, txt, (50, 400), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2.imshow("Task 4: Draw shapes and text", img_4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
