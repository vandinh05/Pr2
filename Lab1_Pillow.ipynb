Task 2: Change color spaces (RBG -> Grayscale, HSV, LAB)

```python
img_2 = Image.open("input_Pillow.jpg")

img_gray = img_2.convert("L")
img_hsv = img_2.convert("HSV")
img_lab = img_2.convert("LAB")

display(img_gray) 
display(img_hsv) 
display(img_lab)
```

Task 3: Crop and resize images

```python
img_3 = Image.open("input_Pillow.jpg")

crop = img_3.crop((100, 100, 400, 400))
display(crop)
```

```python
resize1 = img_3.resize((200, 200))

resize2 = img_3.resize((300, 400), Image.LANCZOS)

# auto caculate the new height to maintain the ratio
w, h = img_3.size
new_w = 400
new_h = int(h * (new_w / w))
resize3 = img_3.resize((new_w, new_h), Image.BICUBIC)

display(resize1, resize2, resize3)
```

Task 4: Draw shape and text on images

```python
from PIL import ImageDraw, ImageFont

img_4 = Image.open("input_Pillow.jpg")
draw = ImageDraw.Draw(img_4)

draw.line([10, 10, 200, 200], fill=(255, 0, 0), width=3)
draw.rectangle([50, 50, 250, 180], outline=(0, 255, 0), width=3)
draw.ellipse([250, 250, 350, 350],fill=(0, 0, 255), outline=(0, 0, 255), width=3)

display(img_4)
```
